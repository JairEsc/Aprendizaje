```python
import polars as pl
```

```python
polar=pl.read_csv("familia.csv")
polar.head()
```

```python
polar.select(pl.col('CAMDA23_MetaSUB_gCSD17_TOK_10'))
```

```python
new_array=polar.select(pl.col('CAMDA23_MetaSUB_gCSD17_TOK_10').alias('Tokio_10'))
```

```python
polar=polar.with_columns((pl.col('CAMDA23_MetaSUB_gCSD17_TOK_10')==0).alias('Una Categorica'))
```

```python
polar.columns
```

```python
polar.group_by('Una Categorica').agg(
    [
        pl.mean('CAMDA23_MetaSUB_gCSD17_TOK_13').alias('mean_tokio_10')
    ]
)
```
