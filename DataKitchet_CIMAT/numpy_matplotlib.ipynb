```python
# prompt: read a .txt and save it as a matrix with float entries

import numpy as np
matrix = np.loadtxt('dataA.txt', dtype=float)
matrix[:,1]
```

```python
# prompt: use matplotlib  to plot a scatter plot with the rows of 'matrix'. Remove axis labels but keep axis

import matplotlib.pyplot as plt

plt.plot(matrix[:,0],matrix[:,1],'o')
```

```python
# prompt: linear regression given X and Y using numpy

import matplotlib.pyplot as plt
import numpy as np
m, b = np.polyfit(matrix[:,0], matrix[:,1], 1)
plt.plot(matrix[:,0],matrix[:,1],'o')
plt.plot(matrix[:,0], m*matrix[:,0] + b)
plt.show()
```
